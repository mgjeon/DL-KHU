# 2D Laplace Equation - Analytical Solution

- https://youtu.be/2aJ3fFwET68

The Laplace equation for unknown function $u$ is

$$
\nabla^2 u = 0
$$

In 2D Cartesian coordinates $(x, y)$,

$$
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0
$$

[Separation of variables](https://en.wikipedia.org/wiki/Separation_of_variables)

$$
u(x, y) = v(x)w(y)
$$

Then the Laplace equation becomes

$$
v_{xx} w + v w_{yy} = 0
$$

Set a constant as $\lambda$

$$
\frac{v_{xx}}{v} = -\frac{w_{yy}}{w} = \lambda
$$

We get two linear homogeneous ordinary differential equations.

$$
\begin{aligned}
v_{xx} - \lambda v & = 0 \\
w_{yy} + \lambda w & = 0 
\end{aligned}
$$

---

If $\lambda=0$, 

$$
\begin{aligned}
v(x) & = c_1 x + c_2 \\
w(y) & = c_3 y + c_4
\end{aligned}
$$


If $\lambda\ne 0$, 

$$
w(y) = c_1 e^{\sqrt{-\lambda}y} + c_2 e^{-\sqrt{-\lambda}y}
$$

When $\lambda < 0$, $\lambda = -\omega^2$, $w(y) = c_1 e^{\omega y} + c_2 e^{-\omega y}$,

$A=c_1+c_2, B=c_1-c_2$

$$
w(y) = c_1 e^{\omega y} + c_2 e^{-\omega y} = A\cosh(wy) + B\sinh(wy)
$$

When $\lambda > 0$, $w(y) = c_1 e^{i\sqrt{\lambda} y} + c_2 e^{-i\sqrt{\lambda} y}$,

$\lambda=\omega^2$, $A=c_1+c_2, B=i(c_1-c_2)$

$$
w(y) = c_1 e^{i\sqrt{\lambda} y} + c_2 e^{-i\sqrt{\lambda} y} = A\cos(\omega y) + B\sin(\omega y)
$$

---

If $\lambda = -\omega^2 < 0$,

$$
\begin{aligned}
v(x) & = \cos(\omega x), \sin(\omega x) \\
w(y) & = \cosh(\omega y), \sinh(\omega y)
\end{aligned}
$$


If $\lambda = 0$,

$$
\begin{aligned}
v(x) & = 1, x \\
w(y) & = 1, y
\end{aligned}
$$

If $\lambda = \omega^2 > 0$,

$$
\begin{aligned}
v(x) & = \cosh(\omega x), \sinh(\omega x) \\
w(y) & = \cos(\omega y), \sin(\omega y)
\end{aligned}
$$

---

- PDE: $\displaystyle \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0$
- Domain: $x \in [0, a], y \in [0, b]$
- Boundary conditions: $\displaystyle u(x, 0) = 0, u(x, b)=\frac{\sin(\omega x)}{\sin(\omega a)}, u(0, y) = 0, u(a, y) = \frac{\sinh(\omega y)}{\sinh(\omega b)}$

Then, the solution is

$$
u(x, y) = \frac{\sin(\omega x)}{\sin(\omega a)}\frac{\sinh(\omega y)}{\sinh(\omega b)}
$$

```python
import numpy as np 
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
# Setup
a = 3*np.pi
b = np.pi
nx = 100
ny = 50
x = np.linspace(0, a, nx)
y = np.linspace(0, b, ny)
```

```python
# Analytical solution
x, y = np.meshgrid(x, y)
u = (np.sin(x)/np.sin(a)) * (np.sinh(y)/np.sinh(b))
u.shape
```

```python
# 2D
fig = plt.figure()
ax = fig.add_subplot(111)
cf = ax.pcolormesh(x, y, u/np.max(u), cmap='seismic')
cbar = fig.colorbar(cf, ax=ax)
cbar.set_label('$u/u_{max}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(0, a)
ax.set_ylim(0, b)
plt.show()
```

```python
# 2D
fig = plt.figure()
ax = fig.add_subplot(111)
cf = ax.contourf(x, y, u/np.max(u), cmap='seismic')
cbar = fig.colorbar(cf, ax=ax)
cbar.set_label('$u/u_{max}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(0, a)
ax.set_ylim(0, b)
plt.show()
```

```python
# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, u/np.max(u), cmap='seismic')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$u/u_{max}$')
ax.set_box_aspect((1,1,1), zoom=0.9)
plt.show()
```
