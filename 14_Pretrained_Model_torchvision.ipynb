```python
def print_conv(model):
    i = 1
    for layer in model.features:
        name = layer.__class__.__name__
        if "Conv" in name:
            print(i, name)
            i += 1
```

```python
from urllib.request import urlretrieve

#goldfish
url = "https://github.com/EliSchwartz/imagenet-sample-images/blob/master/n01443537_goldfish.JPEG?raw=true"
urlretrieve(url, "goldfish.jpg");
```

```python
from PIL import Image
Image.open("goldfish.jpg")
```

```python
from torchvision.io import decode_image
img = decode_image("goldfish.jpg")
img.shape
```

```python
weights = models.VGG16_Weights.DEFAULT
weights
```

```python
preprocess = weights.transforms()
preprocess
```

```python
batch = preprocess(img).unsqueeze(0)
batch.shape
```

```python
model = models.vgg16(weights=None)

prediction = model(batch).squeeze(0)
print("prediction.shape: ", prediction.shape)
print("prediction.sum(): ", prediction.sum())

probability = prediction.softmax(0)
print("probability.sum(): ", probability.sum())

class_id = prediction.argmax().item()
print("class_id: ", class_id)

score = probability[class_id].item()
print("score: ", score)

category_name = weights.meta["categories"][class_id]
print("category_name: ", category_name)
```

```python
model = models.vgg16(weights=weights)

prediction = model(batch).squeeze(0)
print("prediction.shape: ", prediction.shape)
print("prediction.sum(): ", prediction.sum())

probability = prediction.softmax(0)
print("probability.sum(): ", probability.sum())

class_id = prediction.argmax().item()
print("class_id: ", class_id)

score = probability[class_id].item()
print("score: ", score)

category_name = weights.meta["categories"][class_id]
print("category_name: ", category_name)
```
